```python
california['Start_Time'] = pd.to_datetime(california['Start_Time'], format='%Y-%m-%d %H:%M:%S', dayfirst=True, errors='coerce')
def fahrenheit_to_celsius(fahrenheit):
  """Converts Fahrenheit to Celsius"""
  return (fahrenheit - 32) * 5 / 9

california['Temperature(C)'] = california['Temperature(F)'].apply(fahrenheit_to_celsius)
california['Wind_Chill(C)'] = california['Wind_Chill(F)'].apply(fahrenheit_to_celsius)
california["Pressure(cm)"] = [ temp * 2.54 for temp in california["Pressure(in)"]]
def miles_to_km(miles):
  """Converts miles to kilometers"""
  return miles * 1.609

california["Visibility(km)"] = california["Visibility(mi)"].apply(miles_to_km)
california["Precipitation(cm)"] = [ temp * 2.54 for temp in california["Precipitation(in)"]]
california['Wind_Speed(kph)'] = [ mph * 1.609 for mph in california['Wind_Speed(mph)']]
```

```python
target = california["Severity"].to_numpy()
categories = ['Start_Lat', 'Start_Lng', 'Humidity(%)', 'Temperature(C) ',
       'Wind_Chill(C) ', 'Pressure(cm)', 'Visibility(km)', 'Month', 'Hour', 'Wind_Speed(kph)', 'Precipitation(cm)']
input_columns = california[categories].fillna(0).to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_columns, target, test_size=0.2)
```

```python
# k = 500 #number of neighbors to look at (look at 50 nearest neighbors)
# model = KNN(n_neighbors=k)
# model.fit(x_train, y_train)
parameters = {'C':[1, 50], 'epsilon':[0.01,10.0]}
svr = GridSearchCV(SVR(), parameters)
model = make_pipeline(StandardScaler(), svr)

model.fit(x_train, y_train)

# model_score = model.score(x_test, y_test)
```

```python
y_hat = model.predict(x_test)
y_hat
```

```python
total_squared_error = np.sum((y_test - y_hat)**2) #get the sum of all the errors (error = what we want (y_test) - what we predicted (y_hat))
mean_squared_error = total_squared_error/len(y_test) #divide this by how many rows/observations we have
print(str(100.0 - mean_squared_error * 100.0) + str("% accuracy"))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predicted severity on the test set (already calculated in your code)
y_hat = model.predict(x_test)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_hat))

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_hat)

# R-squared
r2 = r2_score(y_test, y_hat)

# Mean Absolute Percentage Error (MAPE)
# Note: This can be undefined for zero actual severity values
mape = np.mean(np.abs((y_test - y_hat) / y_test)) * 100  # Calculate as percentage

# Print the metrics
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared:", r2)
print("Mean Absolute Percentage Error (MAPE):", mape)
```
